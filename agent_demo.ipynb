```python
# agent waiting loop for command with exiting

from recorder import VoiceRecorder
from phrase_to_function import PhraseToFunction
import time

EXIT_COMMANDS = {"exit", "quit", "stop"}

def say(text: str):
    nemo_model.text_to_speech(text, play=True)

sample_commands = {
    "say hi": lambda: say("Hi Arthur"),
    "move forward": lambda: say("Yes! Moving forward"),
    "turn left": lambda: say("Yes! Turning left"),
    "turn right": lambda: say("Yes! Turning right"),
}



def interaction_loop():
    recognizer = PhraseToFunction(sample_commands)
    record = VoiceRecorder()
    say("Say the command for the robot")

    while True:
        try:
            record.record()
            audio = record.get_audio()

            if len(audio) == 0:
                #say("I didn't hear anything.")
                continue

            transcription = nemo_model.transcribe_from_audio(audio)
            user_command = transcription.text.strip().lower()
            print(f"üó£ Command: '{user_command}'")

            if any(exit_word in user_command for exit_word in EXIT_COMMANDS):
                say("Exiting. Goodbye.")
                break

            do_recognize = recognizer.execute(user_command)

            if do_recognize is False:
                print('\a', 'bip') #bip

        except Exception as e:
            print(f"‚ùå Error: {e}")
            say("An error occurred. Try again.")
            time.sleep(1)

if __name__ == "__main__":
    interaction_loop()
```

```python
# sub agent sample

from agent import Agent
from recorder import VoiceRecorder
from phrase_to_function import PhraseToFunction
import time

# Example TTS using NeMo
def say(text: str):
    nemo_model.text_to_speech("\u200B " + text, play=True)  # Padding avoids cutoff

# Microphone, Input callback: returns transcribed text
recorder = VoiceRecorder()

def asr_input():
    recorder.record()
    audio = recorder.get_audio()
    if len(audio) == 0:
        return ""
    return nemo_model.transcribe_from_audio(audio).text.strip().lower()



# Commands for main agent
def get_main_commands(sub_agent_func):
    return {
        "say hi": lambda: say("Hi Arthur"),
        "move forward": lambda: say("Yes! Moving forward"),
        "turn left": lambda: say("Yes! Turning left"),
        "turn right": lambda: say("Yes! Turning right"),
        "register": sub_agent_func,
    }


def sub_agents_closed():
    say("You're back in the main menu. Say a new command.")


EXIT_COMMANDS = {"exit", "quit", "stop"}

# Logic callback for the agent
def agent_logic(commands, input_callback):
    recognizer = PhraseToFunction(commands)
    say("Say the command for the robot")
    while True:
        try:
            command = input_callback()
            if not command:
                continue
            print(f"üó£ Command: '{command}'")

            if command in EXIT_COMMANDS:
                say("Exiting. Goodbye.")
                break

            recognized, result = recognizer.execute(command)
            if not recognized:
                print('\a', "beep")
            if result:
                sub_agents_closed()
                
        except Exception as e:
            print(f"‚ùå Error: {e}")
            say("Error occurred. Try again.")
            time.sleep(1)


def register_agent():   # Sub-agent logic: asks for name and registers it
    def register_logic(commands, input_callback):
        say("What name should I register?")
        name = input_callback()
        if name:
            say(f"{name} registered successfully!")
        else:
            say("I didn't catch a name.")      
    Agent("RegisterAgent", register_logic, asr_input, {}).run()
    return True #return value to the main agent
    



# Main agent setup
main_commands = get_main_commands(register_agent)
main_agent = Agent("MainAgent", agent_logic, asr_input, main_commands)
main_agent.run()
```
